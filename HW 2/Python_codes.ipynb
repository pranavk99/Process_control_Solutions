# Q6. Time series plot of V(x(t)) and X(t)

```python
from scipy.integrate import odeint

def func(x, t):
    dxdt = -(x*x*x)
    return dxdt

t = np.linspace(0,200)
sol = odeint(func, 1, t)

v_x = 0.5*(sol**2)

plt.plot(t, sol, color = 'g', label = 'X(t)')
plt.plot(t, -sol,color = 'g')
plt.plot(t, v_x, color = 'r', label = 'V(x(t))')
plt.xlabel('Time (s)')
plt.ylabel('V(x(t)) and X(t)')
plt.title('Comparison Plot')
plt.legend();
```
