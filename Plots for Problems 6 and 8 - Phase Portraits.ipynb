# Problem 6: Original form

Part 1.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(X, t):
    x1, x2 = X
    return [x2, -2*x1 - 3*x2]

x1 = np.linspace(-5.0, 5.0, 15)
x2 = np.linspace(-5.0, 5.0, 15)

X1, X2 = np.meshgrid(x1, x2)

t = 0

u, v = np.zeros(X1.shape), np.zeros(X2.shape)
NI, NJ = X1.shape

for i in range(NI):
    for j in range(NJ):
        x = X1[i, j]
        y = X2[i, j]
        xprime = f([x, y], t)
        u[i,j] = xprime[0]
        v[i,j] = xprime[1]
        
Q = plt.quiver(X1, X2, u, v, color='g')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
xaxis = np.linspace(-5.0, 5.0, 15)
yaxis = np.linspace(-5.0, 5.0, 15)
zero_vector = np.zeros(15)
plt.plot(xaxis,zero_vector, color = 'r')
plt.plot(zero_vector, xaxis, color = 'r')
```
