## 1) The Lyapunov based control  

```python
# Lyapunov control
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def func(y, t):
    c = 1
    u = -(1+c)*y
    dydt = np.power(y,3)  + (u*np.power(y,2))
    return dydt

# Time
t = np.linspace(0,20)

# Initial condition
x0 = 1
#solution
sol1 = odeint(func,x0, t)

# Initial condition
x0 = -1
#solution
solm1 = odeint(func,x0, t)

# Initial condition
x0 = 5
#solution
sol5 = odeint(func,x0, t)

# Initial condition
x0 = -5
#solution
solm5 = odeint(func,x0, t)

#plot results
plt.title("Closed Loop performance (c = 1)")
plt.plot(t,sol1, label = 'x0 = 1', color = 'pink', linewidth = '3')
plt.plot(t,solm1, label = 'x0 = -1')
plt.plot(t,sol5, label = 'x0 = 5')
plt.plot(t,solm5, label = 'x0 = -5')
plt.xlabel("Time")
plt.ylabel("x (system dynamics)")
plt.legend()
plt.grid()
```

## 2) Sontag Based Control

```python
# Sontag control
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def func(y, t):
    #c = np.sqrt(1+np.power(y,4))
    if y !=0:
        u = -(y + y*np.sqrt(1 + y**4))
    else:
        u = 0
    dydt = np.power(y,3)  + u*(np.power(y,2))
    return dydt

# Time
t = np.linspace(0,20)

# Initial condition
x0 = 0.1
#solution
sol1 = odeint(func,x0, t)

# Initial condition
x0 = -1
#solution
solm1 = odeint(func,x0, t)

# Initial condition
x0 = 2
#solution
sol5 = odeint(func,x0, t)

# Initial condition
x0 = -5
#solution
solm5 = odeint(func,x0, t)



#plot results
plt.title("Closed Loop performance (c = 1)")
plt.plot(t,sol1, label = 'x0 = 1', linewidth = '1')
plt.plot(t,solm1, label = 'x0 = -1', color = 'r', linewidth = '1.5')
plt.plot(t,sol5, label = 'x0 = 5')
plt.plot(t,solm5, label = 'x0 = -5', color = 'g')
plt.xlabel("Time")
plt.ylabel("x (system dynamics)")
plt.legend()
plt.grid()
```
