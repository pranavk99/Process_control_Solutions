```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def lyup(x, c):
    u = -x*(c+1)
    return(u)

def sonta(x, c):
    if x != 0:
        u = -x + x*(np.sqrt(1+np.power(x,4)))
    else:
        u = 0
    return(u)
```

```python
x = np.linspace(0,5,10)
u_lyp = np.zeros(len(x))
u_sonta = np.zeros(len(x))

for i in range(0,len(x)):
    u_lyp[i] = lyup(x[i], 3)
    u_sonta[i] = sonta(x[i],np.sqrt(1+x[i]**4))
    
plt.title("Lyapunov vs Sontag")
plt.plot(abs(x),abs(u_lyp), label = 'lyapunov');
plt.plot(abs(x),abs(u_sonta), label = 'Sontag');
plt.xlabel('|x|')
plt.ylabel('|u|')
plt.xlim(0,5)
plt.ylim(0,120)
plt.legend();
plt.savefig("Lyapu_vs_sontag.jpeg", format = 'jpeg')
```
